# **TEXT CLEANING**

CONVERTING TEXT INTO LOWER CASE

```python
df['text']=df['text'].apply(lambda x: x.lower())
```

```python
df
```

REMOVING PUNCTUATIONS

```python
import string

def remove_punc(txt):
  return txt.translate(str.maketrans('','',string.punctuation))
```

```python
df['text']=df['text'].apply(remove_punc)
```

REMOVING NUMBERS

```python
def remove_nums(txt):
  new = ""
  for i in txt:
    if not i.isdigit():
      new = new + i
  return new
```

```python
df['text']=df['text'].apply(remove_nums)
```

REMOVING EMOJIS AND SPECIAL CHARACTERS

```python
def remove_emojis(txt):
  new=""
  for i in txt:
    if i.isascii():
      new +=i
  return new

df['text']=df['text'].apply(remove_emojis)
```

REMOVE STOPWORDS

```python
import nltk
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
nltk.download('punkt')  #for tokenization
nltk.download('stopwords')
```

```python
stop_words= set(stopwords.words('english'))
```

```python
len(stop_words)
```

```python
df.loc[1]['text']
```

```python
def remove(txt):
  words= word_tokenize(txt)
  cleaned = []
  for i in words:
    if not i in stop_words:
      cleaned.append(i)
  return ' '.join(cleaned)
```

```python
nltk.download('punkt_tab')
```

```python
df['text']=df['text'].apply(remove)
```

```python
df.loc[1]['text']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df['text'], df['emotion'], test_size=0.20, random_state=42 )
```

```python
X_train
```

```python
df.shape
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
```

**BOW IMPLEMENTATION**

```python
bow_vectorizer = CountVectorizer()
```

```python
X_train_bow = bow_vectorizer.fit_transform(X_train)
X_test_bow = bow_vectorizer.transform(X_test)
```

```python
X_train_bow
```

```python
X_test_bow
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score
```

Using naive bayes Model

```python
nb_model = MultinomialNB()
```

```python
nb_model.fit(X_train_bow, y_train)
```

```python
pred_bow = nb_model.predict(X_test_bow)
```

```python
print(accuracy_score(y_test,pred_bow))
```

**TF-IDF IMPLEMENTATION**

```python
tfidf_vectorizer = TfidfVectorizer()
```

```python
X_train_tfidf = tfidf_vectorizer.fit_transform(X_train)
X_test_tfidf = tfidf_vectorizer.transform(X_test)
```

Using naive bayes Model

```python
nb2_model = MultinomialNB()
```

```python
nb2_model.fit(X_train_tfidf, y_train)
```

```python
pred_tfidf = nb2_model.predict(X_test_tfidf)
```

```python
print(accuracy_score(y_test,pred_tfidf))
```

Using Logistic Regression Model

```python
from sklearn.linear_model import LogisticRegression
```

```python
log_model = LogisticRegression(max_iter=1000)
```

```python
log_model.fit(X_train_tfidf, y_train)
```

```python
pred_tfidf = log_model.predict(X_test_tfidf)
```

```python
print(accuracy_score(y_test,pred_tfidf))
```

**ACCURACY IMPROVED WHEN LOGISTIC REGRESSION MODEL IS USED**
